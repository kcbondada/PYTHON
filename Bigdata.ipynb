```python
import pandas as pd
from sqlalchemy import create_engine

# Sample DataFrames (replace these with your actual DataFrames)
df_population = pd.DataFrame({'Country': ['USA', 'China', 'India'],
                              'Year': [2010, 2010, 2010],
                              'Population': [308745538, 1337705000, 1230984507]})

df_gdp = pd.DataFrame({'Country': ['USA', 'China', 'India'],
                       'Year': [2010, 2010, 2010],
                       'GDP': [14992052, 6098700, 1827636]})

# Connect to the SQLite database (replace 'sqlite:///your_database.db' with your database URL)
engine = create_engine('sqlite:///your_database.db')

# Load DataFrames into SQL database
df_population.to_sql('population', con=engine, if_exists='replace', index=False)
df_gdp.to_sql('gdp', con=engine, if_exists='replace', index=False)
```
